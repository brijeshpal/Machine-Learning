```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# one features like f1 is univariate analysis
# two features like f1, f2 is bivariate analysis

a=sns.load_dataset("tips")
a.head()
```

# 
Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discreate dimensions or event types. it is very important in Feature Selection

```python
# -1 to +1 = correlation coeficient or pearson coefficient

# Correlation is find out in integer or floating or Numerical data 
# not find out in categorical data 

#a.corr()
```

```python
numeric_data = a.select_dtypes(include=['float64', 'int64']) 
```

```python
sns.heatmap(numeric_data.corr())
```

# Jointplot

A join plots allows to study the relationship between 2 numeric variables. The central chart display their correlation. it is usually a scatterplot, a hexbin plot,or a 2D density plot

Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=a,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=a,kind='reg')
```

# Pair plot

A Pair plot is also know as a scatterplot, in which one variable in the same data row is matched with another variables value, like this:Pairs plots are just elaborations on this, showing all variables paired with all the other variables 

```python
sns.pairplot(a)
```

```python
sns.pairplot(a,hue='sex')

# hue= sex, smoker ,,,jitne colomn name data set me ho whi h
```

# Dist plot

 dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(a['tip'])
```

```python
sns.distplot(a['tip'],kde=False,bins=10)

# kde=kernal density eastimation
```

# Categorical Plots

Seaborn also helps us in doing the analysis on Categorical data points.

boxplot,
violinplot,
countplot,
bar plot

```python
# count plot= either gives x value or y value 

sns.countplot(x='sex',data=a)  # x axis pr banega 
# for y axis sns.countplot(y='sex',data=a)
```

```python
## bar plot = both values given x and y

sns.barplot(x='total_bill',y='sex',data=a)
```

# Box plot

A box and whisker plot(Sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('sex','total_bill',data=a)

# percentile concept ,,,dots are outliar
```

```python
sns.boxplot(data=a,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x='total_bill',y='day',hue='smoker',data=a)
```

# Violin plot

Violin plot hepls us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',hue='smoker',data=a,plette='rainbow')
```
